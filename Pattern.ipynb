Right Triangle

```python
n = 5 # Number of rows for the pattern

for i in range(1, n + 1):
    print("* " * i)
```

pyramid


```python
n = 5  # Number of rows for the pyramid

for i in range(1, n + 1):
    # Print leading spaces
    print(" " * (n - i), end="")
    # Print stars
    print("* " * i)
```

inverted pyramid

```python
n = 5  # Number of rows for the inverted pyramid

for i in range(n, 0, -1):
    # Print leading spaces
    print(" " * (n - i), end="")
    # Print stars
    print("* " * i)
```

diamond

```python
n = 5  # Number of rows for the top half of the diamond

# Top half of the diamond (pyramid)
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("* " * i)

# Bottom half of the diamond (inverted pyramid, excluding the middle row)
for i in range(n - 1, 0, -1):
    print(" " * (n - i), end="")
    print("* " * i)
```

pyramid using numbers

```python
n = 5  # Number of rows for the pyramid

for i in range(1, n + 1):
    # Print leading spaces
    print(" " * (n - i), end="")
    # Print numbers
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

right triangle using numbers

```python
n = 5  # Number of rows for the right triangle

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

butterfly

```python
n = 5  # Half-width of the butterfly

# Upper half of the butterfly
for i in range(1, n + 1):
    print("*" * i, end="")
    print(" " * (2 * (n - i)), end="")
    print("*" * i)

# Lower half of the butterfly
for i in range(n, 0, -1):
    print("*" * i, end="")
    print(" " * (2 * (n - i)), end="")
    print("*" * i)
```

cat

```python
n = 5 # Base size, affects overall dimensions

# Head and ears
print(" " * (n + 1) + "/\_/\ ")
print(" " * (n) + "( o.o )")
print(" " * (n) + "> ^ <")

# Body and Tail
for i in range(1, n - 1):
    if i == 1:
        print(" " * (n) + "  | |__") # Start of body with tail
    elif i == n - 2:
        print(" " * (n) + " (__/)") # Feet/bottom
    else:
        print(" " * (n) + "  | |")

print(" " * (n + 2) + "U U")
```
